## Looping Statements

A looping statement is a series of commands supplied by the programmer that are repeatedly performed until the intended encoded conditions are met. This is mostly utilized in the programming profession to save time and reduce errors.

### While

While loop is a looping statement that focuses on repeating a specific block of code with unknown variables until it reaches the desired conditions, which may or may not be the case depending on the given condition.

```python
## while loops
i, j = 0, 10
while(i<=j):
    print(f"{1}\t|\t{j}")
    i += 1
```

### For

It's the absolute opposite of the while loop. The for loop is a looping statement that repeats a particular piece of code using predefined variables until the required conditions are fulfilled.

```python
# for(int = 0; i<=10; i++){
# printf(i)
# }

i = 0
for i in range(11):
  print(i)
```

```python
playlist = ["Bahay Kubo", "Magandang Kanta", "Buko"]
print('Now Playing:\n')
for song in playlist:
    print(song)
```

## Flow Control

One of the programming languages is flow control. These are the requests made by the program, the instructions and statements that are performed when the program is running, and the sequence in which they are made. It also determines which code in the program should be executed at a certain moment.

### Condition Statements

Conditional statements, as the name of the programming language suggests, are statements that programmers use to make decisions depending on a particular condition. Unless there is a condition that must be satisfied and may change as a result of the program, it runs the program smoothly.

```python
num_1, num_2 = 14, 12
if(num_1 == num_2):
  print("HAHA")
elif(num_1>num_2):
  print("HOHO")
else:
  print("HUHU")
```
